## Model Validation

```python
# Basic or Naive approach for model Validation

from sklearn.datasets import load_iris

iris = load_iris()

X = iris.data

y = iris.target
```

```python
from sklearn.neighbors import KNeighborsClassifier

model = KNeighborsClassifier(n_neighbors= 1)
```

```python
model.fit(X, y)
```

```python
y_model= model.predict(X)
```

```python
from sklearn.metrics import accuracy_score

accuracy_score(y, y_model)
```

### Model Validation using Holdout sets(Test Set)

```python


from sklearn.model_selection import train_test_split

X1, X2, y1, y2 = train_test_split(X, y, train_size = 0.5, random_state = 0)
```

```python
# fit the model on one set of data
model.fit(X1, y1)
```

```python
# evaluate the model on the second set of data

y2_model = model.predict(X2)
```

```python
y2_model
```

```python
# Model evaluation for Classification  Models
accuracy_score(y2, y2_model) # Correctly Classified Samples / Total number of Sample
```

### Model Validation via Cross-Validation

- Sequence of fits where each subset of the data is used both as a training set and as a validation set

```python
y2_model = model.fit(X1, y1).predict(X2) #test set

y1_model = model.fit(X2, y2).predict(X1) #cross-validation
```

```python
accuracy_score(y1, y1_model)
```

```python
accuracy_score(y2, y2_model)
```

```python
from sklearn.model_selection import cross_val_score

cross_val_score(model, X, y, cv=5)
```

## The bias-variance trade-off

$R^2$ - which is called as the **coefficient of determination**, which measures how well a model performs relative 
to a simple mean(average) of the target values

- $R^2$ = 1, indicates a perfect match
- $R^2$ = 0, indicates the model does no better than taking the mean of the data, negative values mean even worse models

- For high-bias models, the performance of the model on the validation set is similar to the performance on the training set(low)

- For high-variance models, the performance of the model on the validation set is far worse than the performance on the training set
