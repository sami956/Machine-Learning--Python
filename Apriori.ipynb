```python
# Importing the Libraries

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Importing the dataset

data = pd.read_csv('grocery_store.csv', header=None)
data
```

```python
!pip install apyori
```

```python
transactions = []

for i in range(0,20):
    transactions.append([str(data.values[i,j]) for j in range(0,4)])
```

```python
from apyori import apriori
```

```python
rules = apriori(transactions, min_support=0.1, min_confidence=0.5, 
                min_lift=2, min_length=2, max_length=2 )
```

```python
2/20
```

```python
1/2
```

```python
results = list(rules)
```

```python
results
```

```python
results[0][2][0][2]
```

```python
def extract(results):
    antecendents = [tuple(result[2][0][0])[0] for result in results]
    consequents =  [tuple(result[2][0][1])[0] for result in results]
    supports    =  [result[1] for result in results]
    confidences =  [result[2][0][2] for result in results]
    lifts       =  [result[2][0][3] for result in results]
    
    return list(zip(antecendents, consequents, supports, confidences, lifts))
```

```python
list1 = [1,2]
list2 = [2,3]

list(zip(list1,list2))
```

```python
resultdf = pd.DataFrame(extract(results), 
                        columns=['antecendents', 'consequents', 'supports', 'confidences', 'lifts'])
```

```python
resultdf.sort_values('lifts', axis=0, ascending=False)
```
